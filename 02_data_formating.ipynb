```python
# Remove ";"
def remove_semicolon(df):
    for col in df.select_dtypes(include=['object', 'string']).columns:
        if df[col].astype(str).str.contains(';').any():
            df[col] = df[col].str.replace(';', ',', regex = False)
    return df
```
